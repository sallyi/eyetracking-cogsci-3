```python
emo_or_functional = ['emo', 'functional']
zero_or_1 = [0,1]
def get_zero_or_one(x):
    return random.choice(zero_or_1)

get_zero_or_one(1)
```

```python
columns = ['emo-primer','functional-primer', 'image-name','target/filler/crosshair', 'primer_1','primer_2']
```

```python
def shuffle_all(df):
    df['0_or_1'] = df['Emotion?'].apply(get_zero_or_one)
    df['primer_1'] = df['0_or_1'].apply(lambda x: emo_or_functional[x])
    df['primer_2'] = df['0_or_1'].apply(lambda x: emo_or_functional[not(x)])
    df = df.sample(frac=1).reset_index(drop=True)
    i = 0
    df = df[columns]
    #print(df_all)
    while i < len(df):
        #print(i)
        line = crosshairs.sample(frac=.25)
        while (df.iloc[i-1]['target/filler/crosshair'] == 'target') and (df.iloc[i]['target/filler/crosshair'] == 'target'):
            print('two targets in a row')
            df = pd.concat([df.iloc[:i], df.iloc[i:].sample(frac=1)]).reset_index(drop=True)
        else:
            print('exiting: ', df.iloc[i-1]['target/filler/crosshair'], df.iloc[i]['target/filler/crosshair'])
            df = pd.concat([df.iloc[:i], line, df.iloc[i:]]).reset_index(drop=True)
        #print(df_all)
        i+=2
    return df

counter_balancing_df = shuffle_all(df[df['target/filler/crosshair'] != 'crosshair'])
counter_balancing_df[counter_balancing_df['target/filler/crosshair'] != 'crosshair']['target/filler/crosshair']
```

```python

counter_balancing_df['emo-primer-id'] = counter_balancing_df['image-name'].apply(lambda x: 'primer-emo-' + x )
counter_balancing_df['funct-primer-id'] = counter_balancing_df['image-name'].apply(lambda x: 'primer-funct-' + x )
counter_balancing_df
```

```python
def get_primer_sent(image, primer_set):
    primer_set = "primer_" + str(primer_set)
    primer_type = image[primer_set]
    if primer_type == 'emo':
        return image['emo-primer']
    elif primer_type == 'functional':
        return image['functional-primer']
    else:
        return None
    
    
def get_primer_id(image, primer_set):
    primer_set = "primer_" + str(primer_set)
    primer_type = image[primer_set]
    if primer_type == 'emo':
        return image['emo-primer-id']
    elif primer_type == 'functional':
        return image['funct-primer-id']
    else:
        return None
```

```python
letter_pairs = [('C','D'), ('E','F'), ('G','H'),('I','J'),('K','L')]
```

```python
import re
```

```python
def shuffle_all_rows(df):
    df['is_target'] = df['image-name'].apply(lambda x: False if re.match("[0-9]", x[-1]) else True)
    df = df.sample(frac=1).reset_index(drop=True)
    while (df.iloc[0]['is_target'] == True):
        df = df.sample(frac=1).reset_index(drop=True)
    i = 1
    while i < len(df):
        while (df.iloc[i-1]['is_target'] == True) and (df.iloc[i]['is_target'] == True):
            print('two targets in a row')
            df = pd.concat([df.iloc[:i], df.iloc[i:].sample(frac=1)]).reset_index(drop=True)
        else:
            print('exiting: ')
            df = pd.concat([df.iloc[:i], df.iloc[i:]]).reset_index(drop=True)
        #print(df_all)
        i+=1
    return df

counter_balancing_df = shuffle_all(df[df['target/filler/crosshair'] != 'crosshair'])
counter_balancing_df[counter_balancing_df['target/filler/crosshair'] != 'crosshair']['target/filler/crosshair']
```

```python
cb_seqs_to_concat = [df_seqs]

for letter_1, letter_2 in letter_pairs:
    print(letter_1, letter_2)
    crosshairs_all = df_seqs[(df_seqs['image-name'].str.contains('__'))][
        df_seqs['Sequence name/ Order'] == 'A'].sample(frac=1).reset_index(
            drop=True).rename(columns={
                "index": "crosshair-index",
                "image-name": "crosshair-name",
            })[["crosshair-index", "crosshair-name"]]
    crosshairs_all.head()

    all_data_primers_need_shuffling = images_primers_df.join(crosshairs_all)

    def randomize_primers(row):
        choice = random.choice(zero_or_1)
        if choice == 1:
            return row
        else:
            row_copy = row.copy()
            row_copy['primer-A'] = row['primer-A']
            row_copy['primer-B'] = row['primer-B']
            return row_copy

    #[row for row in all_data_primers_need_shuffling.to_dict(orient='records')][0:5]
    # shuffle primer assignment while also shuffling all rows
    all_data_primers_shuffled_df = shuffle_all_rows(
        pd.DataFrame([
            randomize_primers(row)
            for row in all_data_primers_need_shuffling.to_dict(
                orient='records')
        ])).reset_index(drop=True)

    seqA_df = pd.DataFrame([{
        k: v
        for k, v in row.items()
        if k in ['image-name', 'crosshair-name', 'primer-A']
    } for row in all_data_primers_shuffled_df.to_dict(orient='records')
                            ]).rename(columns={'primer-A': 'primer'})
    seqA_df.head()

    seqB_df = pd.DataFrame([{
        k: v
        for k, v in row.items()
        if k in ['image-name', 'crosshair-name', 'primer-B']
    } for row in all_data_primers_shuffled_df.to_dict(orient='records')
                            ]).rename(columns={'primer-B': 'primer'})
    seqB_df.head()

    seqAB_dfs = {letter_1: seqA_df, letter_2: seqB_df}

    all_seq_rows = []
    for seq_letter, seq_df in seqAB_dfs.items():
        i = 0
        for seq in seq_df.to_dict(orient='records'):
            all_seq_rows.append({
                'index': i,
                'image-name': seq['primer'],
                'Sequence name/ Order': seq_letter
            })
            i += 1
            all_seq_rows.append({
                'index': i,
                'image-name': seq['crosshair-name'],
                'Sequence name/ Order': seq_letter
            })
            i += 1
            all_seq_rows.append({
                'index': i,
                'image-name': seq['image-name'],
                'Sequence name/ Order': seq_letter
            })
            i += 1
    df_seqs_more = pd.DataFrame(all_seq_rows)
    df_seqs_more.tail(10)
    cb_seqs_to_concat.append(df_seqs_more)
```
